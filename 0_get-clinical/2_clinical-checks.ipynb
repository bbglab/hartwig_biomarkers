#### 0 - read in short and long data 

```r
short <- read.csv(paste0(TMP_DIR,'clinical_short.csv'), header=TRUE, sep = ",", stringsAsFactors=FALSE)
long <- read.csv(paste0( TMP_DIR,'clinical_long.csv'), header=TRUE, sep = ",", stringsAsFactors=FALSE)
```

#### 1 - Sanity check endpoints: PFS, OS, BOR, Relapse

```r
short %>% group_by(Survival_pfs_event, Survival_os_event, Y_best_response_group, Y_relapse, ID_delayed_response) %>% summarise( ct = n())
```

```r
head(short)
```

```r
(short %>% select(Survival_pfs_event, Survival_os_event, ID_time_to_last_measurement_in_days, Survival_time_to_pfs_event, Survival_time_to_os_event) 
      %>% filter( Survival_os_event == 0, ID_time_to_last_measurement_in_days != Survival_time_to_os_event))
```

#### 2 - Random Patient Checks

```r
random_check <- sample(unique(short %>% pull(patientIdentifier)), 30)
```

```r
i <- 15
(short 
    %>% filter( patientIdentifier == random_check[i]) 
    %>% select(Y_best_response_binary, Y_relapse, Survival_pfs_event, Survival_os_event, Survival_time_to_pfs_event, Survival_time_to_os_event)
)

(long 
    %>% filter( patientIdentifier	 == random_check[i]) 
    %>% select(meta_treatmentStartDate, 
               responseDate,
               meta_deathDate, 
               response,
               progression, 
               os_event,
               time_to_response_in_days,
               time_to_death_in_days)
    %>% arrange(responseDate)
)
```

#### 3 - Sanity check dates

```r
dates <- short %>% select(ID_meta_biopsyDate, ID_meta_deathDate, ID_meta_treatmentStartDate, ID_meta_treatmentEndDate, ID_best_response_date, ID_last_response_date)
date_converter <- apply( dates, 2, as.Date); 
checks <- (data.frame(date_converter) 
               %>% transmute( time_to_treat = ID_meta_treatmentStartDate - ID_meta_biopsyDate,
                              time_to_best = ID_best_response_date - ID_meta_treatmentStartDate, 
                              time_to_death = ID_meta_deathDate - ID_meta_treatmentStartDate))
summary(checks)
```

#### 4 - Sanity check treatments

```r
(short 
     %>% select(clinical_meta_consolidatedTreatmentType, contains("post_contains")) 
     %>% group_by_all() 
     %>% summarise(ct = n())
     %>% arrange(desc(ct))
     %>% filter( ct > 10))
```
